```python
#call the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
#import the dataset
credit = pd.read_csv('default of credit card clients.csv', header =1)
```

```python
#visualize the first records of the dataset
credit.head()
```

```python
#General description of the dataset
credit.describe()
```

```python
#Data types of the dataset
credit.info()
```

```python
#save a new df to modify
creditM = credit.copy()
```

```python
#change data type of: SEX, EDUCATION, MARRIAGE, PAY_0, PAY_2, PAY_3, PAY_4, PAY_5, PAY_6 and default payment next month
creditM['SEX'] = creditM['SEX'].astype('category')
creditM['EDUCATION'] = creditM['EDUCATION'].astype('category')
creditM['MARRIAGE'] = creditM['MARRIAGE'].astype('category')
creditM['PAY_0'] = creditM['PAY_0'].astype('category')
creditM['PAY_2'] = creditM['PAY_2'].astype('category')
creditM['PAY_3'] = creditM['PAY_3'].astype('category')
creditM['PAY_4'] = creditM['PAY_4'].astype('category')
creditM['PAY_5'] = creditM['PAY_5'].astype('category')
creditM['PAY_6'] = creditM['PAY_6'].astype('category')
creditM['default payment next month'] = creditM['default payment next month'].astype('category')
```

```python
#remove the ID column
creditM = creditM.drop('ID', 1)
```

```python
#rename the default payment next month column
creditM = creditM.rename(columns = {"default payment next month":"DefaultPaymentNextMonth"}) 
```

```python
#Data types of the modified dataset
creditM.info()
```

```python
#General description of the modified dataset
creditM.describe()
```

```python
#visualize the first records of the modified dataset
creditM.head()
```

```python
#save a new df to change the labels and generate plots
creditMG = creditM.copy()
```

```python
#Change the name of the labels

creditMG['SEX'] = creditMG['SEX'].apply({1:'Masculino', 2:'Femenino'}.get)
creditMG['DefaultPaymentNextMonth'] = creditMG['DefaultPaymentNextMonth'].apply({0:'No Incumple', 1:'Incumple'}.get)
creditMG['EDUCATION'] = creditMG['EDUCATION'].apply({1:'Postgrado', 2:'Universidad',3:'Secundaria',
                                                     0:'Otro',4:'Otro',5:'Otro',6:'Otro'}.get)
creditMG['MARRIAGE'] = creditMG['MARRIAGE'].apply({1:'Casado', 2:'Soltero',3:'Divorciado',4:'Otro'}.get)

creditMG.head()
```

```python
# Factor plot of 'Default Payment Next Month Status'

Plot1 = sns.catplot('DefaultPaymentNextMonth', data=creditMG, kind='count',margin_titles = 'a')
plt.xlabel("Cumplimiento del proximo pago")
plt.ylabel("Cantidad de personas")
plt.title("Analisis general del cumplimiento de pago")

for i, bar in enumerate(Plot1.ax.patches):
    h = bar.get_height()
    Plot1.ax.text(
        i, # bar index (x coordinate of text)
        h+10, # y coordinate of text
        '{}'.format(int(h)),  # y label
        ha='center', 
        va='center', 
        fontweight='bold', 
        size=14)
```

```python
# Pivot table for 'Limit Bal' vs Default Payment Next Month
scoreLimit = pd.qcut(creditMG["LIMIT_BAL"], 10)
pivotLimit= pd.pivot_table(creditMG, index= ['DefaultPaymentNextMonth', scoreLimit], values= "LIMIT_BAL", aggfunc= 'count')
pivotLimitUnstacked = pivotLimit.unstack('DefaultPaymentNextMonth')
pivotLimitUnstacked
```

```python
# Plot of 'Limit Bal' vs Default Payment Next Month
pivotLimitUnstacked.plot(kind='bar', figsize=(10,8))
plt.title("Cumplimiento del proximo pago por Limite de credito")
```

```python
# Pivot table for 'Age' vs Default Payment Next Month
scoreAge = pd.qcut(creditMG["AGE"], 6)
pivotAge= pd.pivot_table(creditMG, index= ['DefaultPaymentNextMonth', scoreAge], values= "AGE", aggfunc= 'count')
pivotAgeUnstacked = pivotAge.unstack('DefaultPaymentNextMonth')
pivotAgeUnstacked
```

```python
# Plot of 'Age' vs Default Payment Next Month
pivotAgeUnstacked.plot(kind='bar', figsize=(8,6))
plt.title("Cumplimiento del proximo pago por Edad")
```
